##### Que 2) In a quant test of the CAT Exam, the population standard deviation is known to be 100. A sample of 25 tests taken has a mean of 520. Construct an 80% CI about the mean.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Given data
mean = 520
std_dev = 100
sample_size = 25
confidence_level = 0.80

# Calculate the Z-score for the 80% confidence interval
z_score = stats.norm.ppf((1 + confidence_level) / 2)

# Calculate the margin of error
margin_of_error = z_score * (std_dev / np.sqrt(sample_size))

# Confidence Interval
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

# Visualize using seaborn
sns.set(style="whitegrid")
x = np.linspace(lower_bound - 50, upper_bound + 50, 100)
y = stats.norm.pdf(x, mean, std_dev / np.sqrt(sample_size))

plt.figure(figsize=(8, 5))
sns.lineplot(x=x, y=y, label="Sample Distribution", color="blue")
plt.axvline(x=lower_bound, color="red", linestyle="--", label=f'Lower Bound: {lower_bound:.2f}')
plt.axvline(x=upper_bound, color="green", linestyle="--", label=f'Upper Bound: {upper_bound:.2f}')
plt.title('80% Confidence Interval for Sample Mean')
plt.xlabel('Test Scores')
plt.ylabel('Probability Density')
plt.legend()
plt.show()

(lower_bound, upper_bound)
```

Que 3) A car believes that the percentage of citizens in city ABC that owns a vehicle is 60% or less. A sales manager disagrees with this.
He conducted a hypothesis testing surveying 250 residents & found that 170 residents responded yes to owning a vehicle.
a) State the null & alternate hypothesis.
b)At a 10% significance level, is there enough evidence to support the idea that vehicle owner in ABC city is 60% or less.


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given data
p_null = 0.60  # Hypothesized proportion
n = 250  # Sample size
successes = 170  # Number of residents owning a vehicle

# Step 1: Calculate sample proportion
p_sample = successes / n

# Step 2: Calculate standard error
standard_error = np.sqrt((p_null * (1 - p_null)) / n)

# Step 3: Calculate z-statistic
z = (p_sample - p_null) / standard_error

# Step 4: Calculate the p-value
p_value = 1 - norm.cdf(z)

# Step 5: Plotting the distribution and z-value
x = np.linspace(-3, 3, 1000)
y = norm.pdf(x)
sns.lineplot(x=x, y=y)
plt.axvline(z, color='red', linestyle='--')
plt.title(f'Z-statistic: {z:.2f}, P-value: {p_value:.3f}')
plt.show()

z, p_value
```

#### Que 5) In left & right-skewed data, what is the relationship between mean, median & mode?
Draw the graph to represent the same.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Create datasets
right_skewed_data = np.random.exponential(scale=2, size=1000)
left_skewed_data = -1 * np.random.exponential(scale=2, size=1000) + 10

# Plotting
plt.figure(figsize=(12, 6))

for i, data in enumerate([right_skewed_data, left_skewed_data], 1):
    plt.subplot(1, 2, i)
    sns.histplot(data, kde=True)
    plt.axvline(np.mean(data), color='red', linestyle='--', label='Mean')
    plt.axvline(np.median(data), color='green', linestyle='--', label='Median')
    plt.axvline(data[np.argmax(np.bincount(data.astype(int)))], color='blue', linestyle='--', label='Mode')
    plt.title('Right-Skewed Data' if i == 1 else 'Left-Skewed Data')
    plt.legend()

plt.show()
```
